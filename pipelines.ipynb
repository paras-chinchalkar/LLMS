```python
from transformers import pipeline

# Load the summarization pipeline
summarizer = pipeline("summarization", model="facebook/bart-large-cnn")

# Input text
text = """
Large language models like BART and T5 have demonstrated remarkable capabilities in a wide range of NLP tasks,
including summarization, translation, and question answering. They work by pretraining on massive datasets and then
finetuning for specific tasks using supervised learning approaches.
"""

# Generate summary
summary = summarizer(text, max_length=50, min_length=20, do_sample=False)

# Output
print("Summary:", summary[0]['summary_text'])
```

```python
#translation
translator=pipeline('translation_en_to_fr')
result=translator('The data sceintist were truely amazed by the powe  and simplicty of the huggingface pipeline API')
print(result[0]['translation_text'])
```

```python
from transformers import pipeline

# Create the zero-shot classification pipeline
classifier = pipeline('zero-shot-classification')

# Run classification
result = classifier(
    "Hugging face transformer library is amazing",
    candidate_labels=["technology", "sports", "politics"]
)

# Print the result
print(result)
```

```python
#classification
classfier=pipeline('text-classification')
result=classfier('The data sceintist were truely amazed by the powe  and simplicty of the huggingface pipeline API')
print(result)
```

```python
#Text generation
generator=pipeline('text-generation')
result=generator("if there one thing i went you to remember about Hugging pipeline it's")
print(result[0]['generated_text'])
```

```python
#Image generation
from diffusers import StableDiffusionPipeline
import torch

# Load the pre-trained Stable Diffusion model (CPU or GPU)
pipe = StableDiffusionPipeline.from_pretrained(
    "runwayml/stable-diffusion-v1-5",
    torch_dtype=torch.float16
)
pipe = pipe.to("cuda" if torch.cuda.is_available() else "cpu")

# Define your prompt
prompt = "A futuristic city floating above the clouds at sunset"

# Generate the image
image = pipe(prompt).images[0]

# Show or save the image
image.show()  # Opens the image
# image.save("generated_image.png")  # Optionally save it
```
